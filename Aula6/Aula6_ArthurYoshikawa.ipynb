# Equação concorrente

```python
def intersect(a, b):
    plt.title('Intersect')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.plot(a, label='Linear positive')
    plt.plot(b, label='Linear negative')
    plt.axhline(y=0, color='k')
    plt.axvline(x=0, color='k')
    plt.legend()
    plt.show()
    
line1 = []
line2 = []
    
for x1 in range(0,10):
    line1.append(-2*x1 + 3)
    line2.append((x1 + 2)/3)
    
intersect(line1, line2)
```

# Numpy na resolução de sistema linear

```python
a = np.array([[2,1], [1,-3]])
b = np.array([3,-2])
x = np.linalg.solve(a,b)
print("Answer:", x)
print("Answer:", np.allclose(np.dot(a,x),b))
```

# Infinitas Soluções

```python
def infinite(a, b):
    plt.title('Infinite')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.plot(a)
    plt.plot(b)
    plt.axhline(y=0, color='k')
    plt.axvline(x=0, color='k')
    plt.show()
    
line1 = []
line2 = []
    
for x1 in range(0,10):
    line1.append(-2*x1 + 3)
    line2.append(-2*x1 + 3)
    
infinite(line1, line2)
```

# Nenhuma Solução

```python
def nothing(a, b):
    plt.title('Nothing')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.plot(a)
    plt.plot(b)
    plt.axhline(y=0, color='k')
    plt.axvline(x=0, color='k')
    plt.show()
    
line1 = []
line2 = []
    
for x1 in range(0,10):
    line1.append(-2*x1 + 3)
    line2.append(-2*x1 + 1)
    
nothing(line1, line2)
```

# Composição Linear

```python
def image(composicao0, composicao1, composicao2, vec):
    plt.plot([vec[0], 0], [vec[1], 0])
    plt.plot([composicao0[0], 0], [composicao0[1], 0], 'blue')    
    plt.plot([composicao1[0], 0], [composicao1[1], 0])
    plt.plot([composicao2[0], 0], [composicao2[1], 0])
    
    plt.show()

a1  = np.array([2,1])
a2 = np.array([1,-3])
_2a1 = 2*a1
u = _2a1 + a2

image(a1, _2a1, a2, u)
```
